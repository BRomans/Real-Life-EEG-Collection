```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from extract_heatmaps import show_statistics, read_eeg_data,  prepare_images, preview_image, save_images, get_prefix

EEG_PATH = '../Mind-Collector/data/lol_2024-08-04T21-10-29/eeg.csv'
OUT_PATH = './data/heatmaps'

data, markers = read_eeg_data(EEG_PATH)
print("nupmy eeg data:", data.shape)
print("pandas markers column:", markers.shape)
show_statistics(markers)
```

```python
import os

CHANNELS = [1, 2, 3, 4]
DURATION = 20
EVENT = 'death'

imgs = prepare_images(data, CHANNELS, markers, EVENT, DURATION)
print("number of images:", len(imgs))
print("Image shape:", imgs[0].shape)

preview_image(imgs[0], CHANNELS, EVENT)
save_images(imgs, os.path.join(OUT_PATH, EVENT), get_prefix(EEG_PATH))
```
